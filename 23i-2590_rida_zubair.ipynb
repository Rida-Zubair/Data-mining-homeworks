## Part-I: Data Pre-Processing & Analysis

### 1. Load the Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Rida Zubair - customer_orders - customer_orders.csv')

# Display the first few rows
print("First few rows of the dataset:")
df.head()
```

```python
# Check basic information
print("\nBasic information:")
df.info()

print("\nNull counts:")
df.isnull().sum()
```

### 2. Data Cleaning

```python
# Identify and display the total number of duplicate rows
duplicates = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicates}")
df = df.drop_duplicates()
print(f"Dataset shape after removing duplicates: {df.shape}")
```

```python
print("Null values in each column:")
df.isnull().sum()
```

### 3. Handling Missing Values


```python

df['quantity'] = df['quantity'].fillna(df['quantity'].median())

df = df.dropna(subset=['date_time'])

print("Dataset after handling missing values:")
df.isnull().sum()
```

### 4. Dataset Summary

```python
print("Descriptive statistics:")
df.describe()
```

**Key Observations:**
- The dataset has orders for items like Burger, Pizza, Pasta, Sandwich.
- Quantities range from 1 to 3, prices are fixed per item.
- After cleaning, no null values remain.
- Duplicates were removed, reducing the dataset size.

## Part-II: Data Visualization & Feature Engineering

### 5. Date-Time Processing

```python
# Convert date_time to datetime
df['date_time'] = pd.to_datetime(df['date_time'])

# Extract new columns
df['date'] = df['date_time'].dt.date
df['time'] = df['date_time'].dt.time
df['month'] = df['date_time'].dt.month
df['hour'] = df['date_time'].dt.hour
df['weekday'] = df['date_time'].dt.weekday

df.head()
```

### 6. Feature Engineering

```python
# Create hour_span
df['hour_span'] = df['hour'].apply(lambda x: f"{x}–{x+1}")

# Replace month numbers with names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June'}
df['month'] = df['month'].map(month_names)

# Replace weekday numbers with names
weekday_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['weekday'] = df['weekday'].map(weekday_names)

df.head()
```

### 7. Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
# Top 10 most ordered items
top_items = df['item_name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_items.values, y=top_items.index)
plt.title('Top 10 Most Ordered Items')
plt.xlabel('Number of Orders')
plt.ylabel('Item Name')
plt.show()
```

```python
# Number of orders received each month
monthly_orders = df['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_orders.index, y=monthly_orders.values)
plt.title('Number of Orders Received Each Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Count of orders received in each hour_span
hour_span_orders = df['hour_span'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=hour_span_orders.index, y=hour_span_orders.values)
plt.title('Count of Orders Received in Each Hour Span')
plt.xlabel('Hour Span')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of orders by period of the day
def get_period(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['period'] = df['hour'].apply(get_period)
period_orders = df['period'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=period_orders.index, y=period_orders.values)
plt.title('Count of Orders by Period of the Day')
plt.xlabel('Period')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Top 5 items ordered in each period of the day
periods = ['Morning', 'Afternoon', 'Evening', 'Night']
for period in periods:
    period_df = df[df['period'] == period]
    top_items_period = period_df['item_name'].value_counts().head(5)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=top_items_period.values, y=top_items_period.index)
    plt.title(f'Top 5 Items Ordered in {period}')
    plt.xlabel('Number of Orders')
    plt.ylabel('Item Name')
    plt.show()
```
